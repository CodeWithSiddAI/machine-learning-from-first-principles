# Coding Assignment

Create a simple **Student Report Card** program.

Requirements:

* Store a student's name and marks in three subjects.
* Calculate the total and average.
* Print whether the student passed (average ≥ 50).
* Organize your code by writing a function that computes the average.

---

```python
def average(x):
    total = 0
    for i in x:
        total += i
    return total / len(x) 

student_name = "Siddharth"
marks = [80, 85, 90]
average_marks = average(marks)
print(average_marks)
if average_marks >= 50:
    print("Passed")


```

Create a NumPy array representing the heights (in cm) of 10 students.

# Coding Assignment

Using NumPy:

1. Create vectors for the marks of two students in three subjects.
2. Add the vectors.
3. Multiply one vector by `2`.
4. Print all results.
5. Print the shape of each vector.


```python
import numpy as np

# Create a NumPy array representing the heights (in cm) of 10 students
heights = np.array([180, 176, 170, 192, 155, 162, 172, 188, 178, 160])

# 1. Print the array
print("1. Original heights array:")
print(heights)

# 2. Print its shape
print("\n2. Array shape:")
print(heights.shape)

# 3. Add 2 cm to every height (Broadcasting)
heights += 2 

# 4. Print the updated array
print("\n4. Updated heights (after adding 2 cm):")
print(heights)

# 5. Print the tallest height using NumPy
print("\n5. Tallest height:")
print(np.max(heights))
```

# Coding Assignment

Create a dataset of **five students** with these features:

* Hours Studied
* Attendance
* Assignments Completed

Store them in a NumPy matrix called `X`.

Then:

1. Print the matrix.
2. Print its shape.
3. Print the second student's data.
4. Print the attendance column.
5. Add `1` to every assignment count.


```python
import numpy as np 

X = np.array([
    [8, 80, 14],
    [9, 81, 17], 
    [3, 67, 7],
    [5, 58, 10],
    [10, 100, 15]
])

print(X)
print(f"X-shape: {X.shape}")
print(f"Second Student's data: {X[1]}")
print(f"Attendence Column: {X[:, 1]}")

X[:, 2] += 1
print(X)
```
