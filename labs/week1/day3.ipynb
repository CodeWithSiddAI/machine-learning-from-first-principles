## Cost function implementation

```python
import numpy as np
```

```python
# House size (sq ft)
X = np.array([800, 1000, 1200, 1500, 1800])

# House prices ($1000)
y = np.array([150, 180, 210, 260, 300])

w = 0.15
b = 20

def compute_cost(X, y, w, b):
    # find the prediction
    y_hat = w * X + b 
    # find the error
    error = y - y_hat 

    mse = (1/len(X)) * np.sum(error ** 2)
    print(mse)

    # or we can directly use np.mean(error**2)

    cost = (1/(2*len(X))) * np.sum(error**2)
    return cost

cost = compute_cost(X, y, w, b)
print(f"Cost of the function is : {cost}")
```
